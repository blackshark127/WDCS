```python
# ============================================
# 0. IMPORTAR LIBRERÍAS
# ============================================
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

plt.rcParams['figure.figsize'] = (8,4)

# ============================================
# 1. SUBIR ARCHIVO
# ============================================
print("Delapazcoronado_RafaelSaid_Calculos.xlsx")
uploaded = files.upload()

# Tomamos el primer archivo que subas
file_name = list(uploaded.keys())[0]
print(f"Archivo cargado: {file_name}")

# Leemos el excel
df = pd.read_excel(file_name)
print("Primeras filas del dataset crudo:")
display(df.head())

print("\nInformación general del dataset crudo:")
print(df.info())
# ============================================
# 2. LIMPIEZA GENERAL DEL DATASET
# ============================================

# 2.1 Eliminar filas totalmente vacías
df = df.dropna(how='all')

# 2.2 Limpiar texto en columnas categóricas (strip espacios)
columnas_texto = [
    'entidadFederativa',
    'sujetoObligado',
    'nombre',
    'denominacion',
    'cargo',
    'area'
]

for col in columnas_texto:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].str.replace(r'\s+', ' ', regex=True)

# 2.3 Convertir columnas numéricas (texto -> número)
columnas_numericas = ['montoneto', 'montobruto'] # Corrected capitalization

for col in columnas_numericas:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# 2.4 Convertir fechas
columnas_fechas = ['periodoReporteInicio', 'periodoReporteFin']

for col in columnas_fechas:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# 2.5 idInformacion como texto (ID categórico)
if 'idInformacion' in df.columns:
    df['idInformacion'] = df['idInformacion'].astype(str).str.strip()

# 2.6 Eliminar filas donde falten datos en columnas numéricas clave
df = df.dropna(subset=[col for col in columnas_numericas if col in df.columns])

# 2.7 Quitar filas duplicadas completas
df = df.drop_duplicates()

print("Limpieza básica completada.")
print("\nInformación del dataset limpio:")
print(df.info())

print("\nPrimeras filas del dataset limpio:")
display(df.head())
# ============================================
# 3. ESTADÍSTICA DESCRIPTIVA
#    (media, mediana, moda, rango, var, desv)
# ============================================

num_cols_presentes = [col for col in columnas_numericas if col in df.columns]

print("=== DESCRIPTIVO BÁSICO (describe) ===")
display(df[num_cols_presentes].describe())

# Cálculos adicionales
resultados = []

for col in num_cols_presentes:
    serie = df[col].dropna()
    media = serie.mean()
    mediana = serie.median()
    moda = serie.mode().tolist()  # puede haber más de una moda
    minimo = serie.min()
    maximo = serie.max()
    rango = maximo - minimo
    varianza = serie.var()  # muestral
    desv = serie.std()      # muestral

    resultados.append({
        'Variable': col,
        'Media': media,
        'Mediana': mediana,
        'Moda(s)': moda,
        'Mínimo': minimo,
        'Máximo': maximo,
        'Rango': rango,
        'Varianza (muestral)': varianza,
        'Desv. estándar (muestral)': desv
    })

tabla_estadistica = pd.DataFrame(resultados)
print("=== TABLA RESUMEN DE ESTADÍSTICA DESCRIPTIVA ===")
display(tabla_estadistica)
# ============================================
# 4. HISTOGRAMAS
# ============================================

for col in num_cols_presentes:
    df[col].hist(bins=50)
    plt.title(f"Histograma - {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()

# ============================================
# 5. BOXPLOTS
# ============================================

for col in num_cols_presentes:
    df.boxplot(column=[col])
    plt.title(f"Boxplot - {col}")
    plt.show()
# ============================================
# 6. DETECCIÓN DE OUTLIERS CON IQR
# ============================================

outliers_info = []

for col in num_cols_presentes:
    serie = df[col].dropna()
    Q1 = serie.quantile(0.25)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers_col = df[(df[col] < limite_inferior) | (df[col] > limite_superior)]
    num_out = len(outliers_col)

    outliers_info.append({
        'Variable': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Límite inferior': limite_inferior,
        'Límite superior': limite_superior,
        'Número de outliers': num_out
    })

    print(f"\n=== OUTLIERS EN {col} ===")
    print(f"Q1 = {Q1}")
    print(f"Q3 = {Q3}")
    print(f"IQR = {IQR}")
    print(f"Límite inferior = {limite_inferior}")
    print(f"Límite superior = {limite_superior}")
    print(f"Número de outliers: {num_out}")
    print("\nEjemplo de outliers (primeras 5 filas):")
    display(outliers_col.head())

tabla_outliers = pd.DataFrame(outliers_info)
print("\n=== RESUMEN DE OUTLIERS POR VARIABLE ===")
display(tabla_outliers)
# ============================================
# 7. GUARDAR DATASET LIMPIO
# ============================================

output_name = "Dataset_Semana1_Limpio.xlsx"
df.to_excel(output_name, index=False)
print(f"Archivo limpio guardado como: {output_name}")

files.download(output_name)
```

# Celda para Limpiar Excel

```python
import pandas as pd

# Suponemos que df ya está cargado desde el Excel original
df = pd.read_excel("Delapazcoronado_RafaelSaid_Calculos.xlsx")
# =========================
# 1. ELIMINAR FILAS TOTALMENTE VACÍAS
# =========================
df = df.dropna(how='all')

# =========================
# 2. LIMPIEZA DE COLUMNAS DE TEXTO
# =========================
columnas_texto = [
    'entidadfederativa',
    'sujetoobligado',
    'nombre',
    'denominacion',
    'cargo',
    'area'
]

for col in columnas_texto:
    if col in df.columns:
        # Pasar todo a string, quitar espacios extra
        s = df[col].astype(str)
        s = s.str.strip()                       # quita espacios al inicio/fin
        s = s.str.lstrip('-* ')                # quita guiones, asteriscos y espacios al inicio
        s = s.str.replace(r'\s+', ' ', regex=True)  # colapsa espacios múltiples en uno
        df[col] = s

# OPCIONAL: si quieres eliminar filas donde "nombre" tenga números dentro:
if 'nombre' in df.columns:
    mask_nombres_con_numeros = df['nombre'].str.contains(r'\d', na=False)
    # Nos quedamos SOLO con los nombres que NO tienen números
    df = df[~mask_nombres_con_numeros]

# =========================
# 3. LIMPIEZA DE COLUMNAS NUMÉRICAS
# =========================
columnas_numericas = ['montoneto', 'montobruto']

for col in columnas_numericas:
    if col in df.columns:
        # Convertir a número, todo lo raro se vuelve NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

# En este punto, cualquier cosa como "-", "*", "ABC", "12A34" en montoneto/montobruto ya es NaN

# Nos quedamos solo con filas donde montoneto y montobruto:
#  - no sean NaN
#  - sean estrictamente mayores que 0
filtro_validos = pd.Series(True, index=df.index)

if 'montoneto' in df.columns:
    filtro_validos &= df['montoneto'].notna() & (df['montoneto'] > 0)

if 'montobruto' in df.columns:
    filtro_validos &= df['montobruto'].notna() & (df['montobruto'] > 0)

df = df[filtro_validos]

# =========================
# 4. FECHAS E ID
# =========================
if 'periodoreportainicio' in df.columns:
    df['periodoreportainicio'] = pd.to_datetime(df['periodoreportainicio'], errors='coerce')

if 'periodoreportafin' in df.columns:
    df['periodoreportafin'] = pd.to_datetime(df['periodoreportafin'], errors='coerce')

if 'idInformacion' in df.columns:
    df['idInformacion'] = df['idInformacion'].astype(str).str.strip()

# =========================
# 5. QUITAR DUPLICADOS
# =========================
df = df.drop_duplicates()

print("Dataset después de la limpieza por reglas:")
print(df.info())
df.head()
```
